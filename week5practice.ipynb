Calculate mean square loss between predicted changes from previous task and target vector y = [3.0,5.0,8.0]

```python
y = np.array([[3.0],[5.0],[8.0]])

def mean_squared_loss(y, y_hat):
    L = np.mean((y - y_hat)**2)
    return L


loss = mean_squared_loss(y, y_hat)
loss
```

Example algorithm

```python
import numpy as np
import matplotlib.pyplot as plt
n = 100
age = np.linspace(0, 1, n) #age, standardized

thickness = 2.8 - 0.008 * age + np.random.normal(0, 0.05, n)
print(thickness) #cortical thickness with noise

# Training data
X = age
y = thickness
n = len(X)

#Initialize parameters
w = 0.0
b = 0.0
eta = 1e-2 #learning rate
epochs = 2000   #number of iterations

# Gradient Descent
for epoch in range(epochs):
    y_hat = w * X + b
    dw = (2/n) *np.sum((y_hat - y) * X)  #gradient wrt w
    db = (2/n) *np.sum(y_hat - y)         #gradient wrt b

    #update parameters
    w -= eta * dw
    b -= eta * db


    #print progress
    if epoch % 200 == 0:
        loss = np.mean((y_hat - y)**2)
        print(f'Epoch {epoch:4d}: w={w:.3f}, b={b:.3f}, loss={loss:.3f}')

#final weights
print(f"Final weights: w = {w:.3f}, b={b:.3f}")

plt.figure(figsize=(8,6))
plt.scatter(X, y, color="black", label='Data (observed)', alpha=0.7)
plt.plot(age, w*age + b, color="red", label='Fitted regression line', linewidth=2)
plt.xlabel('Age (standardized)', fontsize=14)
plt.ylabel('Cortical Thickness (mm)', fontsize=14) 
plt.title('Linear Regression Fit via Gradient Descent', fontsize=16)
plt.legend(fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()

           
      

```
